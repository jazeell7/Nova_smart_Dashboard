```python
import pandas as pd
import numpy as np
```

```python
n1 = pd.read_csv(r'C:\Users\jaazz\OneDrive\Desktop\P2_NovaMart_Jazeelmuzammil_v1\data\monthly_orders\novamart_orders_2024_01.csv')
n2 = pd.read_csv(r'C:\Users\jaazz\OneDrive\Desktop\P2_NovaMart_Jazeelmuzammil_v1\data\monthly_orders\novamart_orders_2024_02.csv')
n3 = pd.read_csv(r'C:\Users\jaazz\OneDrive\Desktop\P2_NovaMart_Jazeelmuzammil_v1\data\monthly_orders\novamart_orders_2024_03.csv')
n4 = pd.read_csv(r'C:\Users\jaazz\OneDrive\Desktop\P2_NovaMart_Jazeelmuzammil_v1\data\monthly_orders\novamart_orders_2024_04.csv')
n5 = pd.read_csv(r'C:\Users\jaazz\OneDrive\Desktop\P2_NovaMart_Jazeelmuzammil_v1\data\monthly_orders\novamart_orders_2024_05.csv')
n6 = pd.read_csv(r'C:\Users\jaazz\OneDrive\Desktop\P2_NovaMart_Jazeelmuzammil_v1\data\monthly_orders\novamart_orders_2024_06.csv')
customer = pd.read_csv(r'C:\Users\jaazz\OneDrive\Desktop\P2_NovaMart_Jazeelmuzammil_v1\data\customer_lookup.csv')
product = pd.read_csv(r'C:\Users\jaazz\OneDrive\Desktop\P2_NovaMart_Jazeelmuzammil_v1\data\product_lookup.csv')
```

```python
df = pd.concat([n1, n2, n3, n4, n5, n6])
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

```python
df["sales"] = (
    df["sales"]
    .str.replace("$","",regex=False)
    .str.replace(",","",regex=False)
    .astype(float)
)
```

```python
df["order_date"].head(10)
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["ship_date"] = pd.to_datetime(df["ship_date"], errors="coerce")
```

```python
print(df["order_date"].isna().sum())
print(df["ship_date"].isna().sum())
```

```python
df["category"] = df["category"].str.strip().str.title()

df["region"] = df["region"].str.strip().str.title()
```

```python
df["discount"] = df["discount"].fillna(0)
```

```python
df["order_month"] = df["order_date"].dt.month_name()
```

```python
df["profit_margin_pct"] = (
    df["profit"] /
    df["sales"]
) * 100
```

```python
df["shipping_days"] = (
    df["ship_date"] -
    df["order_date"]
).dt.days
```

```python
df = df.merge(
    product[["product_id", "product_name", "unit_cost", "list_price"]],
    on="product_id",
    how="left"
)
```

```python
df = df.merge(
    customer[[
        "customer_id",
        "customer_name",
        "segment",
        "city"
    ]],
    on="customer_id",
    how="left"
)
```

```python
df.to_csv(
    "output/novamart_clean.csv",
    index=False
)
```
